## Import Modules

```python
# Import Module
import os
```

## Set Path

```python
# Set Path
path = '../../content/'
```

## Find All Markdown Files In All Content Folders

```python
# Find all markdown filesin all content folders
all_md_files = [os.path.join(root, name)
    for root, dirs, files in os.walk(path)
        for name in files
            if name.endswith((".md"))]
all_md_files
```
